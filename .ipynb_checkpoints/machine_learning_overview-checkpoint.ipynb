**What is machine learning?**

- Method of data analysis that automates the building of models
- ML iteratively learns from the data to provide new insights without being explicitly programmed to do so

**What is machine learning used for?**

A variety of applications
- Spam filtering, financial modeling, prediction of equipment failtures, fraud detection, text sentiment analysis, etc.

**What does the process look like?**

- Data Acquisition ->
- Data Cleaning ->
- Test Data & Models Training & Building ->
- Models Training & Building -> Model Testing (loop)
- Model deployment

**What are the 3 main types of ml algorithms**

- Supervised Learning
    - Labeled data
    - Predicting a label based off of known features (outputs)
    - Algorithm learns by comparing its outputs with correct outputs to find errors. It then modifies the model accordingly
    - Commonly used in applications where historical data predicts likely future events. 
    - Popular Techniques
        - Classification
        - Regression
        - Prediction and gradient boosting
    - Examples
        - Predicting the price of a house


- Unsupervised Learning
    - Unlabeled data
    - The system is not told the right answer
    - Goal is to explore data and find structures within it
    - Popular Techniques
        - Nearest neighbor mapping
        - k-means clustering


- Reinforcement Learning
    - Algorithm learns to perform an action from experience
    - Often used for robotics, gaming, navigation
    - Through trial and error, algorithm discovers which action yields greatest rewards
    - 3 primary components
        - Agent (learner or decision maker)
        - Environment 
        - Actions
