**What is the prediction algorithm?**


- Calculate the distance from X to all points in data
- Sort points in data by increasing distance to X
- Predict the majority label of the **"k"** closest points

**What happens when the K value increases?**

**What are some of the pros of the KNN algorithm?**


- Ease of use
- Training is very straightforward
- Works with any number of classes
- Easy to add more data
- Few Parameters
    - K
    - Distance Metric

**What are some of the cons?**


- High prediction cost - worse for large datasets
- Does not do well with categorical data
