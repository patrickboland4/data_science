**Where is/are the Root, Nodes, Edges, and Leaves of the screenshot below?**

- Root = outlook

    - The root occurs at the position before a split has taken place
- Nodes = outlook, humidity, windy

    - Nodes split for the value of a certain attribute
    - They are, in essence, our **features**
- Edges = sunny, rain overcast, <=75, >=75, Yes, No

    - These are the outcomes of a split to the next node
    - They are the **observations** contained within the feature column

**What problem do Decision Trees present and why use Random Forests?**

- Decision trees have poor predictive accuracy
    - Susceptible to high variance
    - Different splits in the training data can lead to different trees
- Random Forests provide us with an ensemble of decision trees
    - A new random sample of features is chosen for **each tree at each split**
