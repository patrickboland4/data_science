## Overview of K Means Clustering

#### What are the primary characteristics of K Means Clustering?

- Unlabeled data
- Unsupervised machine learning

#### What does the algorithm do?

- Divide data into distinct groups such that observations within each group are similar

#### What are some examples of relevant applications?

- Cluster similar documents
- Cluster customers based on similar features
- Market segmentation
- Identifying similar physical groups

#### How does the algorithm work?

- Choose a number of clusters **"K"**
- Randomly assign each point to a cluster
- Until clusters stop changing, repeat the following:
    - For each cluster, compute the cluster centroid by taking the mean vector of points in the cluster
    - Assign each data point to the cluster for which the centroid is closest

#### Is there an easy way to choose the K value?

- Not so much
- But there is the **elbow method**
    - Compute the sum of squared error (SSE) for different values of K (ex, 2, 4, 6, 8, etc.)
    - SSE defined as the sum of the squared distance between each member of the cluster and its centroid

```python
Image(filename = PATH + "k_means2.png", height=450, width=450)
```

#### Explaining the elbow method with the picture above

- Above, we have plotted K (x-axis) against SSE (y-axis)
- As K grows larger, our SSE becomes smaller. This is because as the number of clusters increases, the model experiences less distortion
- **Notice the 'elbow' formation when K ~ 6**
    - We see that our gains by increasing K do not yield a significant decrease beyond K ~ 6
